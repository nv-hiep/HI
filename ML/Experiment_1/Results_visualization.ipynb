```python
data = curve.to_numpy(copy=True)
data = data[:, 1:]
```

## Increase dimensionality using Polynomial features (concatenate in 0st axis)

This method using formula x3 = b + (x2)^2 to calculate the third row and concatenate by 0st axis.
In here, b is the index positional embedding, x2 is the value of spectra in each velocity.

```python
data = Increase_Dimension_0.to_numpy(copy=True)
data = data[:, 1:]
```

```python
train_MSE_Indexadd = data[:, 0]
validation_MSE_Indexadd = data[:, 1]
Testing_MSE_Indexadd = data[:, 2]
```

```python
plt.figure(figsize=(7,5))
plt.title('Increase dimensionality using Polynomial features (concatenate in 0st axis)')
plt.plot(train_MSE_Indexadd, label='Training_MES')
plt.grid()
plt.plot(validation_MSE_Indexadd, label='Vaidation_MSE')
plt.grid()
plt.plot(Testing_MSE_Indexadd, label='Tesing_MSE')
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.grid()
plt.legend()
plt.show()
```
