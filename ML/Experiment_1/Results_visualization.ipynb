# Data visualization

## packages

```python
import seaborn as sns
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import sys
```

```python
print(os.getcwd())
```

```python
sns.set(style = "whitegrid")
#sns.set_theme()("whitegrid", {'axes.grid' : False})
```

## Load data from csv

```python
csv_path =  os.getcwd()+"\\training_process\\"
```

```python
concatenate_in_1axis_experiment = csv_path + "concatenate_in_1axis.csv"
Index_info_Add_experiment = csv_path+ "Index_info_Add.csv"
curve_experiment = csv_path+ "curve.csv"
spectrogram_experiment = csv_path+ "spectragram.csv"
No_positional_info = csv_path+ "no_positional_info.csv"
```

```python
Index_info_concatenate_experiment_data = pd.read_csv(concatenate_in_1axis_experiment)
Index_info_Add_experiment_data = pd.read_csv(Index_info_Add_experiment )
curve_experiment_data = pd.read_csv(curve_experiment)
spectrogram_experiment_data = pd.read_csv(spectragram_experiment)
no_positional_info_experiment_data = pd.read_csv(No_positional_info)
```

## Conactenate index embedding experiment

```python
data = Index_info_concatenate_experiment_data.to_numpy(copy=True)
data = data[:, 1:]
```

```python
train_MSE_Indexconcatenate = data[:, 0]
validation_MSE_Indexconcatenate = data[:, 1]
Testing_MSE_Indexconcatenate = data[:, 2]
```

```python
plt.figure(figsize=(7,5))
plt.title('Conactenate index embedding experiment results')
plt.plot(train_MSE_Indexconcatenate, label='Training_MES')
plt.grid()
plt.plot(validation_MSE_Indexconcatenate, label='Vaidation_MSE')
plt.grid()
plt.plot(Testing_MSE_Indexconcatenate, label='Tesing_MSE')
plt.grid()
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.grid()
plt.legend()
plt.show()
```
