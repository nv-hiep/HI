## Load data from csv

```python
csv_path =  os.getcwd()+"\\training_process\\"
```

```python
mse_dict = {}
for i in range (0, 17):
    if i>=2:
        if i%2==0:
            mse_dict[i] = pd.read_csv(csv_path + f"num_layer_{i}.csv")
print(len(mse_dict))
```

## calculate mean 

```python
tmean = {}
vmean = {}
def get_test_mean(dictt):
    for i in range (0, 17):
        if i>=2:
            if i%2==0:
                csv = dictt[i].to_numpy()
                vali_mean = np.mean(csv[:, 1])
                test_mean = np.mean(csv[:, 2])
                tmean[i] = test_mean
                vmean[i] = vali_mean
```

```python
get_test_mean(mse_dict)
```

## Display 

```python
def plott(dictt, title):
    plt.figure(figsize=(7,5))
    plt.title(f'{title} $(10^{(-3)})$')
    p = [2,4,6,8,10, 12, 14, 16]
    x = []
    for i in p:
        value = dictt[i]*1000
        x.append(value)
    plt.plot(p,x,marker='o')
    for xy in zip(p, x):
        plt.annotate('(%.0f, %.2f)' % xy, xy=xy)
    plt.xticks(p)
    plt.grid()
    plt.xlabel('# layers')
    plt.ylabel('MSE mean')
    plt.grid()
    plt.show()
```

```python
plott(tmean, title='mean of test MSE')
```

```python
plott(vmean, title='mean of validation MSE')
```

```python
def plott_both(dictt1,dictt2, title):
    plt.figure(figsize=(7,5))
    plt.title(f'{title} $(10^{-3})$')
    p = [2,4,6,8,10, 12, 14, 16]
    x = []
    for i in p:
        value = dictt1[i]*1000
        x.append(value)
    plt.plot(p,x,marker='o', label = 'mean of test MSE')
    x = []
    for i in p:
        value = dictt2[i]*1000
        x.append(value)
    plt.plot(p,x,marker='o',  label = 'mean of vali MSE')
    plt.xticks(p)
    plt.grid()
    plt.legend()
    plt.xlabel('# layers')
    plt.ylabel('MSE mean')
    plt.grid()
    plt.show()
```

```python
plott_both(tmean, vmean, title='MSE mean in test set')
```

```python
def swarmplot_test(dictt):
    plt.figure(figsize=(9,5))
    A = []
    labelss = []
    for x,y in dictt.items():
        y= y.to_numpy(copy=True)
        y = y[:,2]
        A.append(y)
        labelss.append(x)
    ax = sns.swarmplot(A)
    ax.grid()
    ax.set_xticklabels(labelss, rotation=0)
    plt.title(f'testing MSE swarm plot')
    plt.ylabel('MSE')
    plt.xlabel('# of layers')
    plt.show()
```

```python
swarmplot_test(mse_dict)
```

```python
def get_mean_std(x):
    return np.mean(x), np.std(x)
```

```python
print("validation performance")
for x,y in mse_dict.items():
    y = y.to_numpy(copy=True)
    validation_mse = y[:,1]
    test_mse = y[:,2] 
    m1 ,s1 = get_mean_std(validation_mse)
    m2 ,s2 = get_mean_std(test_mse)
    print(f" mean={m1:.6f}, std={s1:.6f}  for {x}")
```

```python
print("testing performance")
for x,y in mse_dict.items():
    y = y.to_numpy(copy=True)
    test_mse = y[:,2] 
    m1 ,s1 = get_mean_std(test_mse)
    print(f" mean={m1:.6f}, std={s1:.6f}  for {x}")
```
